# 4. Relational Database Fundamentals

## Tables
Store data in rows and columns.

Example:

| id | name | age | email |
|----|------|-----|-------|


## Rows (Records)
Each row represents one entity instance.

Example:
One row equals one customer.

## Columns (Fields)
Each column represents an attribute.

Example:
CustomerID, Name, Email.

## Primary Key (PK)

- Unique identifier for each record
- No duplicates allowed
- Usually ends with “ID” or “Number”

Example:
customerID

## Foreign Key (FK)

- A column that links to a primary key in another table
- Creates relationships between tables

Example:
customerID inside Orders table

## Retrieving All Information from a Table

## Retrieving a Subset of Columns

Instead of *, specify column names separated by commas

````python
%%sql
SELECT lastName, firstName
FROM employees;
````

## Using Aliases (AS)

Aliases rename columns in the output

````python
%%sql
SELECT firstName AS name
FROM employees;
````

AS name → Renames firstName to name in the result

## Using SQL CASE Statements

The CASE statement allows conditional logic inside SQL queries.

Basic Structure
Simple CASE
``` sql
CASE expression
    WHEN value1 THEN result1
    WHEN value2 THEN result2
    ELSE default_result
END
```

````python
%%sql
SELECT firstName, lastName, jobTitle,
       CASE
           WHEN jobTitle = "Sales Rep" THEN "Sales Rep"
           ELSE "Not Sales Rep"
       END AS role
FROM employees;
````

## Shorter CASE Syntax

If comparing one column repeatedly:

````python
%%sql
SELECT firstName, lastName, officeCode,
       CASE officeCode
           WHEN "1" THEN "San Francisco, CA"
           WHEN "2" THEN "Boston, MA"
           WHEN "3" THEN "New York, NY"
           WHEN "4" THEN "Paris, France"
           WHEN "5" THEN "Tokyo, Japan"
       END AS office
FROM employees;
````

##  Built-in SQL Functions

SQL includes built-in functions for:

- String manipulation

- Mathematical operations

- Date and time operations

## String Functions
`length()`

Returns the number of characters in a string.

````python
%%sql
SELECT length(firstName) AS name_length
FROM employees;
````

`upper()`

Converts text to uppercase.

````python
%%sql
SELECT upper(firstName) AS name_in_all_caps
FROM employees;
````

`substr()`

Extracts part of a string.

````python
%%sql
SELECT substr(firstName, 1, 1) || "." AS first_initial
FROM employees;
````

`Concatenation`

Combine columns using ||

%%sql
SELECT firstName || " " || lastName AS full_name
FROM employees;

`CAST()`

## Introduction to the WHERE Clause

The WHERE clause filters rows based on a condition.

The simplest condition checks whether a column equals a specific value using =.

````python
%%sql
SELECT *
FROM employees
WHERE lastName = "Patterson";
````

This returns all columns, but only rows where lastName is "Patterson".

## Filtering Columns and Rows Together

````python
%%sql
SELECT firstName, lastName, email
FROM employees
WHERE lastName = "Patterson";
````

This filters both:

Columns (firstName, lastName, email)

Rows (only Patterson records)

# Common Conditional Operators in SQL

| Operator   | Meaning                         | Example Usage |
|------------|----------------------------------|---------------|
| `!=`       | Not equal to                     | `age != 18` |
| `>`        | Greater than                     | `age > 30` |
| `>=`       | Greater than or equal to         | `age >= 21` |
| `<`        | Less than                        | `age < 50` |
| `<=`       | Less than or equal to            | `age <= 100` |
| `AND`      | Both conditions must be true     | `age > 18 AND age < 30` |
| `OR`       | At least one condition is true   | `breed = "Tabby" OR breed = "Calico"` |
| `BETWEEN`  | Value within a range (inclusive) | `age BETWEEN 2 AND 5` |
| `IN`       | Match any value in a list        | `breed IN ("Tabby", "Calico")` |
| `LIKE`     | Pattern matching (text search)   | `name LIKE "M%"` |
